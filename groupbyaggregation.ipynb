```python
import numpy as np
import pandas as pd
```

```python
data ={
    "Category" : ['A',"B",'A',"B",'A',"B",'A',"B"],
    "Store"  : ['S1','S1','S2','S2','S1','S2','S2','S1'],
    "Sales": [100,200,150,250,120,100,200,300],
    "Quantity" : [10,15,12,18,8,20,15,25],
    "Date" : pd.date_range('2023-01-01',periods=8)

}
df = pd.DataFrame(data)
```

```python
df
```

group by category and calculate sum of sales 

```python
cat = df.groupby("Category")
cat
for i,v in cat :
    print(i)
    print(v)
```

```python
cat = df.groupby("Category")['Sales'].sum()
cat
```

Groupby stores and calculate sum of sales 

```python
store = df.groupby("Store")['Sales'].sum()
store
```

Groupby muntiple columns 

```python
#category aur store dono ko group kr do
muntiple = df.groupby(["Category", "Store"])['Sales'].sum()
muntiple
```
