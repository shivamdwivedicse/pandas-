```python
import numpy as np
import pandas as pd 
```

**FINDING MISSING DATA**

```python
data ={
    'A' :[1,2,np.nan ,4,5],
    'B' :[np.nan , 2,3,4,5], 
    'C' :[1,2,3, np.nan , np.nan],
    'D' :[1, np.nan , np.nan , np.nan ,5] 
}
df = pd.DataFrame(data)
df

#np.nan means null values 
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

**REMOVING NULL VALUES**

```python
df
```

```python
df.dropna() # every row has atleast one null value so all rows get removed.
```

```python
#but if
data1 ={
    'A' :[1,2,np.nan ,4,5],
    'B' :[1,2,3,4,5], 
    'C' :[1,2,3, np.nan , np.nan],
    'D' :[1, np.nan , np.nan , np.nan ,5] 
}
df1 = pd.DataFrame(data1)
df1
```

```python
df1.dropna()
```

```python
#using threshold it means it have that amount of null values ex:-
df1.dropna(thresh=3)  #hav atleast three non null value
```

**FILLING THE MISSING DATA**)

```python
df1.fillna(0)
```

```python
values = {'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values)
```
