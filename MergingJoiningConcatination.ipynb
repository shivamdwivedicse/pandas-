```python
import numpy as numpy
import pandas as pd
```

**MERGING TWO DATAFRAMES**

```python
#Two dataframes

employees = pd.DataFrame({
    'employee_id': [1,2,3,4,5],
    'name' : ['john','anna','peter','linda','bob'],
    'department': ['HR',"IT",'Finance','IT',"HR"]
})

salaries = pd.DataFrame({
    'employee_id': [1,2,3,6,7],
    "salary": [60000,80000,65000,70000,90000], 
    "bonus": [5000,10000,7000,8000,12000]
})
```

```python
employees
```

```python
salaries
```

```python
#merge 

pd.merge(employees,salaries)
```

```python
pd.merge(employees,salaries ,on="employee_id") # on means on which thing you merging
```

```python
pd.merge(employees,salaries ,on="employee_id" , how ="inner") 
```

```python
pd.merge(employees,salaries ,on="employee_id" , how ="outer") 
```

```python
pd.merge(employees,salaries ,on="employee_id" , how ="left")# base on employee id  
```

```python
pd.merge(employees,salaries ,on="employee_id" , how ="right") #based on salary
```

**CONCATINATION**

```python
df1 = pd.DataFrame({
    'A': ['A0',"A1","A2"], 
    'B': ['B0',"B1","B2"],
    'C': ['C0',"C1","C2"]
})

df2 = pd.DataFrame({
    "A":['A3','A4','A5'], 
    "B":['B3','B4','B5'],
    "C":['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])  #on basis of column 
```

```python
#basis of row
pd.concat([df1,df2],axis =1)
```

**JOINING TWO DATAFRAMES**

```python
df1 = pd.DataFrame({
    'name' : ['alice','bob','charlie']
},index= [1,2,3])


df2 = pd.DataFrame({
    'score' : [85,90,75]
},index= [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```

```python
df1=pd.DataFrame({
    "id": [1,2,3], 
    "name": ['Ram',"Shyam","Aman"]
})
df2=pd.DataFrame({
    'id':[2,3,4],
    'marks':[80,90,85]
})
```

```python
pd.merge(df1,df2 ,on="id")
```

```python
pd.merge(df1,df2 ,on="id" , how="inner")
```

```python
pd.merge(df1,df2 ,on="id" , how="outer")
```

```python
pd.merge(df1,df2 ,on="id" , how="left")
```

```python
pd.merge(df1,df2 ,on="id" , how="right")
```
